# Chapter 1: What is EDA?

Exploratory Data Analysis (EDA) is the very first thing you do when you get a dataset. Before building any model, before running any statistics — you explore the data. Think of it like a detective inspecting a crime scene before drawing any conclusions.

The term was coined by statistician **John Tukey in 1977**. His idea was simple:
> *"Let the data speak to you before you force a theory on it."*

In plain language — EDA answers questions like:

-  What does this data look like?
-  Is anything missing or broken?
-  Are there patterns, trends, outliers?
-  What are the relationships between variables?

---

> *"Let me show you the big picture first:"*

##  The Big Picture of EDA

| Step | Name | Details |
|------|------|---------|
| 1️ | **Load Data** | Read CSV, Excel, DB |
| 2️ | **First Inspection** | shape, dtypes, info |
| 3️ | **Data Quality** | nulls, duplicates |
| 4️ | **Univariate** | 1 variable at a time |
| 5️ | **Bivariate** | 2 variables together |
| 6️ | **Multivariate** | many variables |
| 7️ | **Visualize** | plots, charts, heatmaps |
| 8️ | **Insights** | patterns, hypotheses |
| 9️ | **Feature Eng.** | prepare for models |

>  Flow: Load → Inspect → Quality → Univariate → Bivariate → Multivariate → Visualize → Insights → Feature Eng.

# Types of Data

**Data** branches into 3 types:

| Type | Description | Sub-types / Examples |
|------|-------------|----------------------|
| **Numerical** | Numbers with meaning | ↳ **Continuous** (height, salary) <br> ↳ **Discrete** (age, count) |
| **Categorical** | Groups, labels | *(no math on these!)* |
| **Ordinal** | Ordered categories | Low < Med < High, 1 – 5 |

---

###  Numerical
- **Continuous** → height, salary *(any value in a range)*
- **Discrete** → age, count *(only whole numbers)*

###  Categorical
- Groups or labels → gender, city, color
-  **No math operations allowed!**

###  Ordinal
- Ordered categories → Low < Medium < High
- Rating scales → 1 to 5

## Understanding Each Data Type

---

###  Numerical — *"Math makes sense here"*
Numbers you can add, subtract, average.

> **Example:** Employee salaries → ₹45,000 | ₹72,000 | ₹1,20,000

| Subtype | Definition | Example |
|---------|------------|---------|
| **Continuous** | Any value in a range | Height = 5.73 ft |
| **Discrete** | Whole counts only | No. of children = 0, 1, 2 |

---

###  Categorical — *"Just labels, no math"*
Groups or names. You **cannot** perform math on them.

> **Examples:**
> - Gender → Male / Female / Other
> - City → Mumbai / Delhi / Hyderabad
> - Product → Electronics / Clothing

>  Male + Female ≠ anything meaningful!

---

###  Ordinal — *"Order matters, but gaps don't"*
Categories with a **natural order**, but unequal gaps between them.

> **Example 1 — Customer Satisfaction:**
> Very Bad → Bad → Neutral → Good → Excellent
> *(You know Excellent > Good, but not by how much)*

> **Example 2 — Education Level:**
> 10th < 12th < Graduate < Postgraduate

##  Chapter 3: Pandas Data Types

In Python's **pandas**, every column has a specific data type.
This matters because pandas **behaves differently** depending on the type.

---

###  Common Pandas Data Types

| Pandas dtype | Meaning | Example |
|-------------|---------|---------|
| `int64` | Whole numbers | age = 25, count = 3 |
| `float64` | Decimal numbers | salary = 45000.50 |
| `object` | Text / strings | name = "Sai", city = "Hyderabad" |
| `bool` | True / False | is_active = True |
| `datetime64` | Dates & times | 2024-01-01 |
| `category` | Categorical labels | gender, product type |

---

###  Critical line — int becomes float!

> When a column has **even one missing value**, pandas upgrades `int64` → `float64`
> because `NaN` is technically a **float** in Python.
